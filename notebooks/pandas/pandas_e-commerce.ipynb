```python
from pathlib import Path

# Project-root relative paths keep the notebook portable across machines.
PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "data").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent.parent

paths = {
    "customers": PROJECT_ROOT / "data" / "olist_customers_dataset.xlsx",
    "geolocation": PROJECT_ROOT / "data" / "olist_geolocation_dataset.xlsx",
    "order_items": PROJECT_ROOT / "data" / "olist_order_items_dataset.xlsx",
    "order_payments": PROJECT_ROOT / "data" / "olist_order_payments_dataset.xlsx",
    "order_reviews": PROJECT_ROOT / "data" / "olist_order_reviews_dataset.xlsx",
    "orders": PROJECT_ROOT / "data" / "olist_orders_dataset.xlsx",
    "products": PROJECT_ROOT / "data" / "olist_products_dataset.xlsx",
    "sellers": PROJECT_ROOT / "data" / "olist_sellers_dataset.xlsx",
    "category_translation": PROJECT_ROOT / "data" / "product_category_name_translation.xlsx",
}

dataframes = {}
```

```python
dataframes = {name: pd.read_excel(path) for name, path in paths.items()}
```

```python
from datetime import datetime

# Some Excel exports auto-format decimal monetary values as dates.
# Example: 26.08 can become 2026-08-26. Convert those cells back to numeric values.
def excel_date_number_to_float(value):
    if isinstance(value, datetime):
        if value.year == 2026:
            return value.day + value.month / 100
        return (value.year - 2000) + value.month / 100
    return value

numeric_columns = {
    "order_items": ["price", "freight_value"],
    "order_payments": ["payment_value"],
    "products": [
        "product_name_lenght",
        "product_description_lenght",
        "product_photos_qty",
        "product_weight_g",
        "product_length_cm",
        "product_height_cm",
        "product_width_cm",
    ],
    "geolocation": ["geolocation_lat", "geolocation_lng"],
}

for table_name, columns in numeric_columns.items():
    for column in columns:
        if column in dataframes[table_name].columns:
            dataframes[table_name][column] = (
                dataframes[table_name][column]
                .map(excel_date_number_to_float)
                .pipe(pd.to_numeric, errors="coerce")
            )
```

```python
customers = dataframes['customers']
geolocation = dataframes['geolocation']
order_items = dataframes['order_items']
order_payments = dataframes['order_payments']
order_reviews = dataframes['order_reviews']
orders = dataframes['orders']
products = dataframes['products']
sellers = dataframes['sellers']
category_translation = dataframes['category_translation']
```

### Блок 2 — Динаміка продажів
6. Як змінювалась кількість замовлень по місяцях?
7. Чи є сезонність у продажах?

```python
# 6. Як змінювалась кількість замовлень по місяцях?
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_month'] = orders['order_purchase_timestamp'].dt.to_period('M').astype(str)

orders_per_month = (
    orders.groupby('order_month')['order_id']
    .nunique()
    .reset_index(name='orders_count')
    .sort_values('order_month')
)

plt.figure(figsize=(12,5))
sns.lineplot(data=orders_per_month, x='order_month', y='orders_count', marker='o')
plt.title('Кількість замовлень по місяцях')
plt.xlabel('Місяць')
plt.ylabel('Кількість замовлень')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 7. Чи є сезонність у продажах? (Перевірка сезонності: щомісячні закономірності за роками + розподіл за календарними місяцями)
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
df = orders.copy()
df['year'] = df['order_purchase_timestamp'].dt.year
df['month'] = df['order_purchase_timestamp'].dt.month

monthly_by_year = df.groupby(['year', 'month'])['order_id'].nunique().reset_index(name='orders_count')
pivot = monthly_by_year.pivot(index='month', columns='year', values='orders_count').sort_index()

plt.figure(figsize=(12,5))
sns.lineplot(data=pivot, markers=True)
plt.title('Monthly orders by year')
plt.xlabel('Month')
plt.ylabel('Number of orders')
plt.xticks(range(1,13))
plt.legend(title='Year')
plt.tight_layout()
plt.show()

# Boxplot: distribution of monthly order counts across calendar months (all years)
orders_per_month = df.groupby(pd.Grouper(key='order_purchase_timestamp', freq='M'))['order_id'].nunique().reset_index(name='orders_count')
orders_per_month['month'] = orders_per_month['order_purchase_timestamp'].dt.month

plt.figure(figsize=(10,5))
sns.boxplot(data=orders_per_month, x='month', y='orders_count')
plt.title('Distribution of monthly orders by calendar month (all years)')
plt.xlabel('Month')
plt.ylabel('Orders count')
plt.tight_layout()
plt.show()
```

```python
# 8. Візуалізація продажів по місяцях
oi = order_items.copy()
oi['price'] = pd.to_numeric(oi['price'], errors='coerce')
oi['freight_value'] = pd.to_numeric(oi['freight_value'], errors='coerce')

sales = oi.merge(orders[['order_id', 'order_purchase_timestamp']], on='order_id', how='left')
sales['order_purchase_timestamp'] = pd.to_datetime(sales['order_purchase_timestamp'])
sales['order_month'] = sales['order_purchase_timestamp'].dt.to_period('M').astype(str)

monthly_gmv = sales.groupby('order_month')['price'].sum().reset_index(name='gmv')
monthly_orders = sales.groupby('order_month')['order_id'].nunique().reset_index(name='orders_count')
monthly = monthly_gmv.merge(monthly_orders, on='order_month', how='left')
monthly['aov'] = monthly['gmv'] / monthly['orders_count']
monthly['order_month_dt'] = pd.to_datetime(monthly['order_month'] + '-01')
monthly = monthly.sort_values('order_month_dt')

plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.lineplot(data=monthly, x='order_month_dt', y='gmv', marker='o')
plt.title('GMV по місяцях')
plt.xlabel('Місяць')
plt.ylabel('GMV (sum of prices)')
plt.xticks(rotation=45)

plt.subplot(1,2,2)
sns.lineplot(data=monthly, x='order_month_dt', y='aov', marker='o', color='orange')
plt.title('Середній чек (AOV) по місяцях')
plt.xlabel('Місяць')
plt.ylabel('AOV')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 9. Як змінюється GMV (Gross Merchandise Value) з часом?
monthly = monthly.sort_values('order_month_dt').reset_index(drop=True)

# rolling averages
monthly['gmv_3m'] = monthly['gmv'].rolling(3, min_periods=1).mean()
monthly['gmv_6m'] = monthly['gmv'].rolling(6, min_periods=1).mean()

plt.figure(figsize=(12,5))
sns.lineplot(data=monthly, x='order_month_dt', y='gmv', marker='o', label='GMV')
sns.lineplot(data=monthly, x='order_month_dt', y='gmv_3m', label='3-month MA', color='orange')
sns.lineplot(data=monthly, x='order_month_dt', y='gmv_6m', label='6-month MA', color='green')

plt.title('GMV (Gross Merchandise Value) з часом')
plt.xlabel('Місяць')
plt.ylabel('GMV (сума цін)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# додатково: відсоткова зміна GMV по місяцях
monthly['gmv_pct_change'] = monthly['gmv'].pct_change()
print(monthly[['order_month','gmv','gmv_pct_change']].tail(8))
```

### Блок 4 — Аналіз платежів
14. Які способи оплати найпопулярніші?
15. Який середній чек по кожному способу оплати?
16. Який дохід по кожному способу оплати?

```python
# 14. Які способи оплати найпопулярніші?
payment_counts = (
    order_payments['payment_type']
    .value_counts()
    .rename_axis('payment_type')
    .reset_index(name='count')
)
payment_counts['pct'] = payment_counts['count'] / payment_counts['count'].sum()

print(payment_counts)

plt.figure(figsize=(8,4))
ax = sns.barplot(data=payment_counts, x='count', y='payment_type', palette='magma')
plt.title('Найпопулярніші способи оплати')
plt.xlabel('Кількість платежів')
plt.ylabel('Спосіб оплати')

total = payment_counts['count'].sum()
for p in ax.patches:
    width = p.get_width()
    ax.text(width + total*0.01, p.get_y() + p.get_height()/2, f"{int(width)} ({width/total:.1%})", va='center')

plt.tight_layout()
plt.show()
```

```python
# Order-level payment totals by payment type.
# Olist can split one order into multiple payment rows, so aggregate before AOV/revenue analysis.
orders_pay = (
    order_payments
    .groupby(['order_id', 'payment_type'], as_index=False)['payment_value']
    .sum()
    .rename(columns={'payment_value': 'order_total'})
)
```

```python
# 15. Який середній чек по кожному способу оплати?
avg_by_payment = (
    orders_pay.groupby('payment_type')['order_total']
    .mean()
    .reset_index(name='avg_order_value')
    .sort_values('avg_order_value', ascending=False)
)

plt.figure(figsize=(8,4))
ax = sns.barplot(data=avg_by_payment, x='avg_order_value', y='payment_type', palette='viridis')
plt.title('Середній чек по способу оплати')
plt.xlabel('Середній чек (USD)')
plt.ylabel('Спосіб оплати')

xlim = ax.get_xlim()
for p in ax.patches:
    width = p.get_width()
    y = p.get_y() + p.get_height() / 2
    text_color = 'white' if width > (xlim[1] * 0.12) else 'black'
    ax.text(width / 2, y, f"{width:,.2f}", ha='center', va='center', color=text_color, fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 16. Який дохід по кожному способу оплати ?
rev_by_payment = (
    orders_pay
    .groupby('payment_type')
    .agg(orders_count=('order_id', 'nunique'),
         total_revenue=('order_total', 'sum'))
    .reset_index()
    .sort_values('total_revenue', ascending=False)
)
rev_by_payment['pct'] = rev_by_payment['total_revenue'] / rev_by_payment['total_revenue'].sum()

plt.figure(figsize=(10,5))
total_rev = rev_by_payment['total_revenue'].sum()
ax = sns.barplot(data=rev_by_payment, x='total_revenue', y='payment_type', palette='magma')
plt.title('Доход по способу оплати')
plt.xlabel('Total Revenue')
plt.ylabel('Payment Type')

for p, (_, row) in zip(ax.patches, rev_by_payment.iterrows()):
    width = p.get_width()
    ax.text(width + total_rev*0.005, p.get_y() + p.get_height()/2,
            f"{width:,.2f} ({row['pct']:.1%})", va='center')

plt.tight_layout()
plt.show()

print(rev_by_payment)
```

### Блок 5 — Відгуки клієнтів
17. Який середній рейтинг замовлень?

```python
# 17. Який середній рейтинг замовлень? — візуалізація (щомісячна динаміка + розподіл рейтингів)
df_rev = order_reviews.copy()
df_rev['review_creation_date'] = pd.to_datetime(df_rev['review_creation_date'])
df_rev['month'] = df_rev['review_creation_date'].dt.to_period('M').astype(str)

monthly_rev = (
    df_rev.groupby('month')['review_score']
    .agg(avg_score='mean', reviews_count='size')
    .reset_index()
)
monthly_rev['month_dt'] = pd.to_datetime(monthly_rev['month'] + '-01')
monthly_rev = monthly_rev.sort_values('month_dt')

plt.figure(figsize=(14,6))

plt.subplot(2,1,1)
sns.lineplot(data=monthly_rev, x='month_dt', y='avg_score', marker='o')
plt.title('Середній рейтинг замовлень по місяцях')
plt.xlabel('')
plt.ylabel('Середній рейтинг')
plt.ylim(1,5)
plt.xticks(rotation=45)

plt.subplot(2,1,2)
# prepare explicit DataFrame for counts to avoid seaborn column lookup issues
rev_counts = df_rev['review_score'].value_counts().sort_index().rename_axis('review_score').reset_index(name='count')
sns.barplot(data=rev_counts, x='review_score', y='count', palette='rocket')
plt.title('Розподіл рейтингів (кількість відгуків)')
plt.xlabel('Рейтинг')
plt.ylabel('Кількість відгуків')

plt.tight_layout()
plt.show()
```

```python
# 18. Яка частка негативних відгуків (1–2)?
df_reviews = order_reviews.copy()

total_reviews = len(df_reviews)
neg_count = df_reviews['review_score'].isin([1, 2]).sum()
neg_pct = neg_count / total_reviews

score_counts = df_reviews['review_score'].value_counts().sort_index()

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.pie([neg_count, total_reviews - neg_count],
    labels=[f'Negative (1-2): {neg_count}', f'Other (3-5): {total_reviews - neg_count}'],
    autopct='%.1f%%', colors=['#d62728', '#2ca02c'], startangle=140)
plt.title('Share of Negative Reviews (scores 1–2)')

plt.subplot(1,2,2)
colors = ['#d62728' if s in [1,2] else '#1f77b4' for s in score_counts.index]
sns.barplot(x=score_counts.index.astype(str), y=score_counts.values, palette=colors)
plt.title('Count of Reviews by Score')
plt.xlabel('Review Score')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# 19. Чи впливає час доставки на рейтинг? — візуалізація
df = order_reviews.merge(
    orders[['order_id','order_purchase_timestamp','order_delivered_customer_date','order_estimated_delivery_date']],
    on='order_id', how='left'
)

df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
df['delivery_days'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days
df = df[df['delivery_days'].notna() & (df['delivery_days'] >= 0)]

bins = [-0.1, 0, 1, 2, 3, 5, 7, 14, 30, 90, 10000]
labels = ['0', '1', '2', '3-4', '5-6', '7-13', '14-29', '30-89', '90-364', '365+']
df['delivery_bin'] = pd.cut(df['delivery_days'], bins=bins, labels=labels)

plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
sns.boxplot(data=df, x='delivery_bin', y='review_score', order=labels)
plt.title('Розподіл рейтингів по інтервалах часу доставки')
plt.xlabel('Час доставки (дні)')
plt.ylabel('Рейтинг')
plt.xticks(rotation=45)

plt.subplot(1,2,2)
mean_by_bin = df.groupby('delivery_bin')['review_score'].mean().reindex(labels)
sns.pointplot(x=mean_by_bin.index, y=mean_by_bin.values, color='orange')
plt.title('Середній рейтинг по інтервалах часу доставки')
plt.xlabel('Час доставки (дні)')
plt.ylabel('Середній рейтинг')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 20. Які категорії товарів отримують найгірші відгуки? — візуалізація
df_cat = (
    order_reviews[['order_id','review_score']]
    .merge(order_items[['order_id','product_id']], on='order_id', how='left')
    .merge(products[['product_id','product_category_name']], on='product_id', how='left')
)

# use english names if available
label_col = 'product_category_name'
if 'product_category_name_english' in category_translation.columns:
    df_cat = df_cat.merge(category_translation, on='product_category_name', how='left')
    label_col = 'product_category_name_english'

# aggregate
cat_scores = (
    df_cat.dropna(subset=['product_category_name'])
    .groupby(label_col)['review_score']
    .agg(avg_score='mean', reviews_count='size')
    .reset_index()
)

# filter noisy categories
min_reviews = 50
cat_scores = cat_scores[cat_scores['reviews_count'] >= min_reviews].sort_values('avg_score').reset_index(drop=True)

# visualize worst categories (bottom 20)
worst = cat_scores.head(20)

plt.figure(figsize=(10,8))
ax = sns.barplot(data=worst, x='avg_score', y=label_col, palette='Reds_r')
plt.title('Worst product categories by average review score (min_reviews={})'.format(min_reviews))
plt.xlabel('Average review score')
plt.xlim(1,5)
for i, row in worst.iterrows():
    ax.text(row['avg_score'] + 0.03, i, f"{row['avg_score']:.2f} ({int(row['reviews_count'])})", va='center')
plt.tight_layout()
plt.show()
```

### Блок 6 — Логістика
21. Скільки в середньому триває доставка?

```python
# 21. Скільки в середньому триває доставка?
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])

df_delivery = orders[['order_id','order_purchase_timestamp','order_delivered_customer_date','order_month']].copy()
df_delivery['delivery_days'] = (df_delivery['order_delivered_customer_date'] - df_delivery['order_purchase_timestamp']).dt.days
df_delivery = df_delivery[df_delivery['delivery_days'].notna() & (df_delivery['delivery_days'] >= 0)]

print(f"Mean delivery days: {df_delivery['delivery_days'].mean():.2f}")
print(f"Median delivery days: {df_delivery['delivery_days'].median():.2f}")

plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
sns.histplot(df_delivery['delivery_days'], bins=40, kde=True)
plt.title('Розподіл часу доставки (дні)')
plt.xlabel('Дні доставки')

plt.subplot(1,2,2)
monthly_avg = (df_delivery.assign(order_month_dt=pd.to_datetime(df_delivery['order_month'] + '-01'))
               .groupby('order_month_dt')['delivery_days'].mean().reset_index())
sns.lineplot(data=monthly_avg, x='order_month_dt', y='delivery_days', marker='o')
plt.title('Середній час доставки по місяцях')
plt.xlabel('Місяць')
plt.ylabel('Середній час доставки (дні)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,5))
df_delivery['month'] = df_delivery['order_purchase_timestamp'].dt.month
sns.boxplot(data=df_delivery, x='month', y='delivery_days')
plt.title('Розподіл часу доставки по календарних місяцях')
plt.xlabel('Місяць')
plt.ylabel('Дні доставки')
plt.tight_layout()
plt.show()
```

```python
# 22. Чи є затримки доставки? — візуалізація
df_delay = orders[['order_id','order_delivered_customer_date','order_estimated_delivery_date','order_month']].copy()
df_delay['order_delivered_customer_date'] = pd.to_datetime(df_delay['order_delivered_customer_date'])
df_delay['order_estimated_delivery_date'] = pd.to_datetime(df_delay['order_estimated_delivery_date'])
df_delay = df_delay.dropna(subset=['order_delivered_customer_date','order_estimated_delivery_date'])
df_delay['delay_days'] = (df_delay['order_delivered_customer_date'] - df_delay['order_estimated_delivery_date']).dt.days
df_delay['delay_status'] = np.where(df_delay['delay_days']>0, 'delayed', 'on_time_or_early')

# overall share
status_counts = df_delay['delay_status'].value_counts()

plt.figure(figsize=(14,4))
plt.subplot(1,3,1)
plt.pie(status_counts.values,
    labels=[f"{s}: {v} ({v/status_counts.sum():.1%})" for s, v in status_counts.items()],
    autopct='%.1f%%', startangle=140, colors=['#d62728','#2ca02c'])
plt.title('Share of delayed vs on-time orders')

# distribution of delay days (only delays)
plt.subplot(1,3,2)
sns.histplot(df_delay.loc[df_delay['delay_days']>0, 'delay_days'], bins=30, color='salmon')
plt.xlabel('Delay days (only delayed)')
plt.title('Distribution of delay days')

# monthly delayed rate
monthly_delay = (df_delay.groupby('order_month')['delay_status']
         .apply(lambda x: (x == 'delayed').mean())
         .reset_index(name='pct_delayed'))
monthly_delay['order_month_dt'] = pd.to_datetime(monthly_delay['order_month'] + '-01')
monthly_delay = monthly_delay.sort_values('order_month_dt')

plt.subplot(1,3,3)
sns.lineplot(data=monthly_delay, x='order_month_dt', y='pct_delayed', marker='o')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))
plt.xticks(rotation=45)
plt.title('Monthly % of delayed orders')
plt.tight_layout()
plt.show()
```

```python
# 23. Які штати мають найбільші затримки?
order_sellers = (
    order_items[['order_id','seller_id']]
    .drop_duplicates()
    .merge(sellers[['seller_id','seller_state']], on='seller_id', how='left')
)

# merge with df_delay (created in previous cell)
df_state = order_sellers.merge(df_delay[['order_id','delay_days','delay_status']], on='order_id', how='inner')
df_state = df_state[df_state['seller_state'].notna()]

state_stats = (
    df_state.groupby('seller_state')
    .agg(orders_count=('order_id','nunique'),
         delayed_count=('delay_status', lambda x: (x=='delayed').sum()),
         avg_delay_days=('delay_days', lambda x: x[x>0].mean()))
    .reset_index()
)

state_stats['pct_delayed'] = state_stats['delayed_count'] / state_stats['orders_count']

top_avg = state_stats.dropna(subset=['avg_delay_days']).sort_values('avg_delay_days', ascending=False).head(10)
top_pct = state_stats.sort_values('pct_delayed', ascending=False).head(10)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.barplot(data=top_avg, x='avg_delay_days', y='seller_state', palette='Reds_r')
plt.title('Top 10 states by average delay days (only delayed)')
plt.xlabel('Average delay days')
plt.ylabel('State')

plt.subplot(1,2,2)
sns.barplot(data=top_pct, x='pct_delayed', y='seller_state', palette='Blues_r')
plt.title('Top 10 states by % of delayed orders')
plt.xlabel('Percent delayed')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0%}'))
plt.ylabel('State')

plt.tight_layout()
plt.show()

# display table for reference
state_stats.sort_values('avg_delay_days', ascending=False).head(20)
```

```python
# 28. Які продавці мають найбільше негативних відгуків? — візуалізація (топ-10)
df_neg = (
    order_reviews[['order_id','review_score']]
    .merge(order_items[['order_id','seller_id']], on='order_id', how='left')
    .dropna(subset=['seller_id'])
)

neg_only = df_neg[df_neg['review_score'].isin([1,2])]

neg_counts = neg_only.groupby('seller_id')['review_score'].count().reset_index(name='neg_count')
total_counts = df_neg.groupby('seller_id')['review_score'].count().reset_index(name='total_reviews')

seller_neg_stats = neg_counts.merge(total_counts, on='seller_id', how='right').fillna(0)
seller_neg_stats['neg_pct'] = seller_neg_stats['neg_count'] / seller_neg_stats['total_reviews']

top_n = 10
top_neg = (
    seller_neg_stats.sort_values(['neg_count','neg_pct'], ascending=[False,False])
    .head(top_n)
    .merge(sellers[['seller_id','seller_state']], on='seller_id', how='left')
)

plt.figure(figsize=(10,6))
ax = sns.barplot(data=top_neg, x='neg_count', y='seller_id', palette='Reds_r')
plt.title(f'Top-{top_n} sellers by number of negative reviews (scores 1–2)')
plt.xlabel('Number of negative reviews')
plt.ylabel('seller_id')

for i, row in top_neg.reset_index().iterrows():
    ax.text(row['neg_count'] + top_neg['neg_count'].max()*0.01, i, f"{int(row['neg_count'])} ({row['neg_pct']:.1%}) — {row['seller_state']}", va='center')

plt.tight_layout()
plt.show()
```

```python
# 29. Які категорії товарів продаються найкраще? — візуалізація (Top-15 by items sold + revenue)
df_cat_sales = sales.merge(products[['product_id','product_category_name']], on='product_id', how='left')

# prefer english labels if available
label_col = 'product_category_name'
if 'product_category_name_english' in category_translation.columns:
    df_cat_sales = df_cat_sales.merge(category_translation, on='product_category_name', how='left')
    label_col = 'product_category_name_english'

# ensure numeric
df_cat_sales['price'] = pd.to_numeric(df_cat_sales['price'], errors='coerce').fillna(0)
df_cat_sales['freight_value'] = pd.to_numeric(df_cat_sales['freight_value'], errors='coerce').fillna(0)
df_cat_sales['revenue'] = df_cat_sales['price'] + df_cat_sales['freight_value']

cat_metrics = (
    df_cat_sales.dropna(subset=[label_col])
    .groupby(label_col)
    .agg(items_sold=('product_id','count'),
         orders_count=('order_id','nunique'),
         revenue=('revenue','sum'))
    .reset_index()
    .sort_values('items_sold', ascending=False)
)

top_n = 15
top_cats = cat_metrics.head(top_n).reset_index(drop=True)

plt.figure(figsize=(12,8))

plt.subplot(1,2,1)
ax = sns.barplot(data=top_cats, x='items_sold', y=label_col, palette='viridis')
plt.title(f'Top {top_n} Categories by Items Sold')
plt.xlabel('Items sold (count)')
plt.ylabel('')
for p in ax.patches:
    ax.text(p.get_width() + top_cats['items_sold'].max()*0.01, p.get_y()+p.get_height()/2, f"{int(p.get_width()):,}", va='center')

plt.subplot(1,2,2)
ax2 = sns.barplot(data=top_cats, x='revenue', y=label_col, palette='magma')
plt.title(f'Top {top_n} Categories by Revenue')
plt.xlabel('Revenue (sum)')
plt.ylabel('')
max_rev = top_cats['revenue'].max()
for p in ax2.patches:
    ax2.text(p.get_width() + max_rev*0.005, p.get_y()+p.get_height()/2, f"{p.get_width():,.0f}", va='center')

plt.tight_layout()
plt.show()
```

```python
# 31. Які категорії мають найгірші відгуки? — візуалізація
df_cat_reviews = (
    order_reviews[['order_id','review_score']]
    .merge(order_items[['order_id','product_id']], on='order_id', how='left')
    .merge(products[['product_id','product_category_name']], on='product_id', how='left')
)

# use english labels if available
label_col = 'product_category_name'
if 'product_category_name_english' in category_translation.columns:
    df_cat_reviews = df_cat_reviews.merge(category_translation, on='product_category_name', how='left')
    label_col = 'product_category_name_english'

cat_scores = (
    df_cat_reviews.dropna(subset=[label_col])
    .groupby(label_col)['review_score']
    .agg(avg_score='mean', reviews_count='size')
    .reset_index()
)

min_reviews = 50
bottom_n = 20
worst_cats = cat_scores[cat_scores['reviews_count']>=min_reviews].sort_values('avg_score').head(bottom_n)

plt.figure(figsize=(10,8))
ax = sns.barplot(data=worst_cats, x='avg_score', y=label_col, palette='Reds_r')
plt.title(f'Найгірші категорії за середнім рейтингом (min_reviews={min_reviews})')
plt.xlabel('Середній рейтинг')
plt.xlim(1,5)
for i, row in worst_cats.reset_index().iterrows():
    ax.text(row['avg_score'] + 0.03, i, f"{row['avg_score']:.2f} ({int(row['reviews_count'])})", va='center')
plt.tight_layout()
plt.show()
```

```python
# 35. Різниця в середньому чеку між регіонами? — візуалізація
order_totals = (
    order_items.assign(item_total=order_items['price'] + order_items['freight_value'])
    .groupby('order_id', as_index=False)['item_total']
    .sum()
    .rename(columns={'item_total': 'order_total'})
)

df_amt = (
    orders[['order_id','customer_id']]
    .merge(customers[['customer_id','customer_state']], on='customer_id', how='left')
    .merge(order_totals, on='order_id', how='left')
)

state_stats = (
    df_amt.dropna(subset=['customer_state','order_total'])
    .groupby('customer_state')['order_total']
    .agg(avg_order='mean', median_order='median', orders_count='count')
    .reset_index()
)

# filter noisy states
min_orders = 50
state_stats_f = state_stats[state_stats['orders_count'] >= min_orders].sort_values('avg_order', ascending=False)

plt.figure(figsize=(10,8))
ax = sns.barplot(data=state_stats_f, x='avg_order', y='customer_state', palette='viridis')
plt.title('Середній чек по штатах (мин. замовлень = {})'.format(min_orders))
plt.xlabel('Середній чек (order_total)')
plt.ylabel('Штат')

# annotate
max_val = state_stats_f['avg_order'].max()
for i, row in state_stats_f.reset_index().iterrows():
    ax.text(row['avg_order'] + max_val*0.01, i, f"{row['avg_order']:.2f} (n={int(row['orders_count'])})", va='center')

plt.tight_layout()
plt.show()
```

```python
# 36. Чи є залежність між відстанню та часом доставки? — візуалізація
# prepare zip centroids
zip_geo = (
    geolocation
    .groupby('geolocation_zip_code_prefix', as_index=False)
    .agg({'geolocation_lat':'median','geolocation_lng':'median'})
    .rename(columns={
        'geolocation_zip_code_prefix':'zip_code_prefix',
        'geolocation_lat':'lat',
        'geolocation_lng':'lng'
    })
)

# map customer and seller zip to orders
order_seller = order_items[['order_id','seller_id']].drop_duplicates('order_id')
df = (
    orders[['order_id','customer_id','order_purchase_timestamp','order_delivered_customer_date']]
    .merge(customers[['customer_id','customer_zip_code_prefix']], on='customer_id', how='left')
    .merge(order_seller.merge(sellers[['seller_id','seller_zip_code_prefix']], on='seller_id', how='left'),
           on='order_id', how='left')
)

# merge lat/lng for customer and seller
df = df.merge(zip_geo.rename(columns={'zip_code_prefix':'customer_zip_code_prefix','lat':'lat_c','lng':'lng_c'}),
              on='customer_zip_code_prefix', how='left')
df = df.merge(zip_geo.rename(columns={'zip_code_prefix':'seller_zip_code_prefix','lat':'lat_s','lng':'lng_s'}),
              on='seller_zip_code_prefix', how='left')

# compute delivery days
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
df['delivery_days'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days

# haversine distance (km)
def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    return 6371 * 2 * np.arcsin(np.sqrt(a))

df_dist = df.dropna(subset=['lat_c','lng_c','lat_s','lng_s','delivery_days']).copy()
df_dist['distance_km'] = haversine(df_dist['lat_c'], df_dist['lng_c'], df_dist['lat_s'], df_dist['lng_s'])
df_dist = df_dist[df_dist['delivery_days'] >= 0]

# plot: scatter + lowess and binned mean
sample = df_dist.sample(n=min(5000, len(df_dist)), random_state=1)

plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
sns.scatterplot(data=sample, x='distance_km', y='delivery_days', alpha=0.25, s=10)
sns.regplot(data=sample, x='distance_km', y='delivery_days', scatter=False, lowess=False, color='red')
plt.xscale('log')
plt.xlabel('Distance (km, log scale)')
plt.ylabel('Delivery days')
plt.title('Distance vs Delivery Days (sample + LOWESS)')

plt.subplot(1,2,2)
bins = [0,1,5,10,20,50,100,200,500,2000]
df_dist['dist_bin'] = pd.cut(df_dist['distance_km'], bins=bins)
mean_by_bin = df_dist.groupby('dist_bin')['delivery_days'].agg(['mean','count']).reset_index().dropna()
sns.barplot(data=mean_by_bin, x='mean', y='dist_bin', palette='viridis')
plt.xlabel('Mean delivery days')
plt.ylabel('Distance bin (km)')
plt.title('Mean delivery days by distance bin')

plt.tight_layout()
plt.show()
```

```python
# 37. Яка частка скасованих замовлень? — візуалізація
df = orders.copy()
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_month'] = df['order_purchase_timestamp'].dt.to_period('M').astype(str)
status = df['order_status'].astype(str).str.lower()

total = len(df)
canceled_mask = status == 'canceled'
canceled_count = canceled_mask.sum()
canceled_pct = canceled_count / total

# Pie + counts
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.pie([canceled_count, total - canceled_count],
    labels=[f"Canceled: {canceled_count}", f"Other: {total - canceled_count}"],
    autopct='%.1f%%', colors=['#d62728','#2ca02c'], startangle=140)
plt.title(f"Share of canceled orders ({canceled_pct:.1%})")

# Monthly cancelled rate
monthly_total = df.groupby('order_month')['order_id'].nunique().reset_index(name='total_orders')
monthly_canceled = df[canceled_mask].groupby('order_month')['order_id'].nunique().reset_index(name='canceled_orders')
monthly = monthly_total.merge(monthly_canceled, on='order_month', how='left').fillna(0)
monthly['canceled_pct'] = monthly['canceled_orders'] / monthly['total_orders']
monthly['month_dt'] = pd.to_datetime(monthly['order_month'] + '-01')
monthly = monthly.sort_values('month_dt')

plt.subplot(1,2,2)
sns.lineplot(data=monthly, x='month_dt', y='canceled_pct', marker='o', color='red')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.1%}'))
plt.title('Monthly % of canceled orders')
plt.xlabel('Month')
plt.ylabel('Canceled %')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 38. Чи пов’язані скасування з певними продавцями? — візуалізація
df_cancel = orders[['order_id','order_status']].copy()
df_cancel['is_canceled'] = df_cancel['order_status'].astype(str).str.lower() == 'canceled'

# ensure one seller per order
order_seller_simple = order_items[['order_id','seller_id']].drop_duplicates('order_id')

df = df_cancel.merge(order_seller_simple, on='order_id', how='left').dropna(subset=['seller_id'])

seller_cancel_stats = (
    df.groupby('seller_id')
      .agg(total_orders=('order_id','nunique'),
           canceled_count=('is_canceled','sum'))
      .reset_index()
)
seller_cancel_stats['canceled_pct'] = seller_cancel_stats['canceled_count'] / seller_cancel_stats['total_orders']

# attach seller state (if available)
seller_cancel_stats = seller_cancel_stats.merge(sellers[['seller_id','seller_state']], on='seller_id', how='left')

# filter noisy sellers
min_orders = 20
seller_f = seller_cancel_stats[seller_cancel_stats['total_orders'] >= min_orders]

# plots
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
sns.scatterplot(data=seller_f, x='total_orders', y='canceled_pct', alpha=0.6)
plt.xscale('log')
plt.xlabel('Total orders (log scale)')
plt.ylabel('Canceled %')
plt.title('Canceled % vs number of orders per seller')

plt.subplot(1,2,2)
top_n = 20
top_by_pct = seller_f.sort_values('canceled_pct', ascending=False).head(top_n)
ax = sns.barplot(data=top_by_pct, x='canceled_pct', y='seller_id', palette='Reds_r')
plt.xlabel('Canceled %')
plt.ylabel('seller_id')
plt.title(f'Top {top_n} sellers by canceled % (min_orders={min_orders})')
for i, row in top_by_pct.reset_index().iterrows():
    ax.text(row['canceled_pct'] + 0.001, i, f"{row['canceled_count']}/{row['total_orders']} ({row['canceled_pct']:.1%})", va='center')

plt.tight_layout()
plt.show()
```

```python
# 39. Чи впливає довга доставка на скасування?
df_cd = orders[['order_id','order_status']].merge(df_delay[['order_id','delay_days']], on='order_id', how='left')
df_cd = df_cd[df_cd['delay_days'].notna()].copy()
df_cd['is_canceled'] = df_cd['order_status'].astype(str).str.lower() == 'canceled'

bins = [-0.1, 0, 1, 3, 7, 14, 30, 90, 1e9]
labels = ['<=0','1','2-3','4-7','8-14','15-30','31-89','90+']
df_cd['delay_bin'] = pd.cut(df_cd['delay_days'], bins=bins, labels=labels, include_lowest=True)
df_cd['delay_bin'] = df_cd['delay_bin'].cat.as_ordered()

stats = df_cd.groupby('delay_bin', observed=True).agg(
    total_orders=('order_id','count'),
    canceled_orders=('is_canceled','sum')
).reset_index()
stats['canceled_pct'] = stats['canceled_orders'] / stats['total_orders']

# plotting
plt.figure(figsize=(11,6))
ax = sns.barplot(data=stats, x='delay_bin', y='canceled_pct', palette='Reds_r', order=labels)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.1%}'))
plt.ylim(0, stats['canceled_pct'].max()*1.2 if stats['canceled_pct'].max()>0 else 0.01)
plt.title('Доля скасованих замовлень за інтервалами затримки доставки')
plt.xlabel('Інтервал затримки (дні)')
plt.ylabel('Доля скасованих замовлень')
plt.grid(axis='y', linestyle='--', alpha=0.3)

# annotate counts and percent
for i, row in stats.iterrows():
    pct = row['canceled_pct']
    ax.text(i, pct + plt.ylim()[1]*0.02, f"{int(row['canceled_orders'])}/{int(row['total_orders'])}\n{pct:.1%}",
            ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# 40. Які категорії мають найбільше проблем? — composite metric (cancellations, negative reviews, delays)
df_cat = order_items[['order_id','product_id']].merge(products[['product_id','product_category_name']], on='product_id', how='left')
label_col = 'product_category_name'
if 'product_category_name_english' in category_translation.columns:
    df_cat = df_cat.merge(category_translation, on='product_category_name', how='left')
    label_col = 'product_category_name_english'

# cancellations per category
df_cancel = orders[['order_id','order_status']].copy()
df_cancel['is_canceled'] = df_cancel['order_status'].astype(str).str.lower()=='canceled'
cat_cancel_stats = (
    df_cat.merge(df_cancel[['order_id','is_canceled']], on='order_id', how='left')
    .dropna(subset=[label_col])
    .groupby(label_col, observed=True)
    .agg(orders_count=('order_id','nunique'),
         canceled_count=('is_canceled', lambda x: x.sum()))
    .reset_index()
)
cat_cancel_stats['canceled_pct'] = cat_cancel_stats['canceled_count'] / cat_cancel_stats['orders_count']

# negative reviews per category
df_rev = order_reviews[['order_id','review_score']].copy()
df_rev['is_neg'] = df_rev['review_score'].isin([1,2])
cat_rev_stats = (
    df_cat.merge(df_rev[['order_id','is_neg']], on='order_id', how='left')
    .dropna(subset=[label_col])
    .groupby(label_col, observed=True)
    .agg(reviews_count=('order_id','count'),
         neg_count=('is_neg', lambda x: x.sum()))
    .reset_index()
)
cat_rev_stats['neg_pct'] = cat_rev_stats['neg_count'] / cat_rev_stats['reviews_count'].replace(0, np.nan)

# delays per category (uses df_delay from earlier cells)
cat_delay_stats = (
    df_cat.merge(df_delay[['order_id','delay_days','delay_status']], on='order_id', how='left')
    .dropna(subset=[label_col])
    .groupby(label_col, observed=True)
    .agg(delay_obs=('delay_days','count'),
         pct_delayed=('delay_status', lambda x: (x=='delayed').mean()),
         avg_delay_days=('delay_days', lambda x: x[x>0].mean()))
    .reset_index()
)

# combine stats
df_stats = (cat_cancel_stats
            .merge(cat_rev_stats[[label_col,'neg_pct','reviews_count']], on=label_col, how='left')
            .merge(cat_delay_stats[[label_col,'pct_delayed','avg_delay_days','delay_obs']], on=label_col, how='left')
           )

df_stats[['neg_pct','reviews_count','pct_delayed','avg_delay_days','delay_obs']] = df_stats[['neg_pct','reviews_count','pct_delayed','avg_delay_days','delay_obs']].fillna(0)

# normalize metrics to [0,1] and sum into problem_score
for col in ['canceled_pct','neg_pct','pct_delayed','avg_delay_days']:
    mx = df_stats[col].max()
    df_stats[col + '_norm'] = df_stats[col] / mx if mx>0 else 0

df_stats['problem_score'] = (df_stats['canceled_pct_norm'] + df_stats['neg_pct_norm'] +
                             df_stats['pct_delayed_norm'] + df_stats['avg_delay_days_norm'])

# filter by activity and show top categories
min_orders = 50
top = (df_stats[df_stats['orders_count'] >= min_orders]
       .sort_values('problem_score', ascending=False)
       .head(20)
       .reset_index(drop=True))

plt.figure(figsize=(10,8))
ax = sns.barplot(data=top, x='problem_score', y=label_col, palette='Reds_r')
plt.title('Top 20 categories by composite problem score (cancellations, negative reviews, delays)')
plt.xlabel('Problem score (normalized sum)')
plt.ylabel('')

for i, row in top.iterrows():
    txt = f"orders={int(row['orders_count'])}  canc={row['canceled_pct']:.1%}  neg={row['neg_pct']:.1%}  delayed={row['pct_delayed']:.1%}  avg_delay={0 if np.isnan(row['avg_delay_days']) else row['avg_delay_days']:.1f}d"
    ax.text(row['problem_score'] + 0.01, i, txt, va='center', fontsize=9)

plt.tight_layout()
plt.show()
```
